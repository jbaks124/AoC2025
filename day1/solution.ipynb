```python
# AoC 2025 - Day 1, Part 1
# Count ending on 0

pos = 50
zero_hits = 0
turns = 0

with open("input.txt") as f:
    for line in f:
        line = line.strip()

        if not line:
            continue

        turns += 1
        
        # First character: ‘L’ of ‘R’
        direction = line[0]

        # Rest of line is clicks
        distance = int(line[1:])

        # Perform turn & wrap around the circle
        if direction == "L":
            pos = (pos - distance) % 100
        else:  # direction == "R"
            pos = (pos + distance) % 100

        # Count how often we land on 0
        if pos == 0:
            zero_hits += 1

print("Zero hits (code):", zero_hits)
print("Total turns:", turns)
```

```python
# AoC 2025 - Day 1, Part 2
# Count passing by 0

pos = 50
zero_hits = 0

with open("input.txt") as f:
    for line in f:
        line = line.strip()
        if not line:
            continue

        direction = line[0]
        distance = int(line[1:])

        # Step direction
        step = -1 if direction == "L" else 1

        # Run steps & count 0
        for _ in range(distance):
            pos = (pos + step) % 100
            if pos == 0:
                zero_hits += 1

print(zero_hits)
```
