```python
# AoC 2025 - Day 2, Part 1
# Walk through ranges, find invalid repeating pattern, and summarize

import re

def is_invalid(number):
    text = str(number)
    pattern = r'(\d+)\1'
    match = re.fullmatch(pattern, text)
    if match is not None:
        return True
    else:
        return False

total = 0

with open("input.txt", "r") as file:
    raw_text = file.read().strip()

range_items = raw_text.split(",")

for item in range_items:
    if item.strip() == "":
        continue

    parts = item.split("-")
    start_id = int(parts[0])
    end_id = int(parts[1])

    for number in range(start_id, end_id + 1):
        if is_invalid(number):
            total += number

print("Answer:", total)
```
