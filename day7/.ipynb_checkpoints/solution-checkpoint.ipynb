```python
g = [l.rstrip('\n') for l in open("input.txt")]
R, C = len(g), len(g[0])
b = [[0]*C for _ in range(R)]

# initial beam
for c,x in enumerate(g[0]):
    if x=="S": b[0][c]=1

splits = 0
for r in range(R-1):
    for c in range(C):
        k = b[r][c]
        if not k: continue
        if g[r][c]=="^":
            splits += 1
            if c>0: b[r+1][c-1]+=k
            if c<C-1: b[r+1][c+1]+=k
        else:
            b[r+1][c]+=k

out=[]
for r in range(R):
    row=""
    for c in range(C):
        row += "|" if b[r][c]>0 else g[r][c]
    out.append(row)

open("solution.txt","w").write("\n".join(out)+f"\n\nSplits: {splits}\n")
print("Splits:", splits)
```
